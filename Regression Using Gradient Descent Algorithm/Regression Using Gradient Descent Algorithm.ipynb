# Regression Using Gradient Descent Algorithm

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Trn_d=pd.read_csv('trainRegression.csv')
Tst_d=pd.read_csv('testRegression.csv')
```

```python
Trn_Int= np.array(Trn_d['X'])
Trn_Out= np.array(Trn_d['R'])
```

```python
Tst_Int=np.array(Tst_d['X'])
Tst_Out=np.array(Tst_d['R'])
```

```python
learning_rates= [0.1 , 0.01 , 0.001]
for learning_rate  in learning_rates:
    Thita=np.zeros((2,1))                      
    new_Thita=np.zeros((2,1))
    Thita[0,0]=random.uniform(0,1)
    Thita[1,0]=random.uniform(0,1)
    new_Thita[0,0]=Thita[0,0]
    new_Thita[1,0]=Thita[1,0]
    Flag=True
    Iterations=0
```

```python
while (not(np.allclose(new_Thita,Thita)) or Flag):
        Flag = False
        Thita[0,0]=new_Thita[0,0]
        Thita[1,0]=new_Thita[1,0]
        
        y_desh_Trn_d = Thita[0,0]+(Thita[1,0] * Trn_Int)
        
        derivative_0 = ( sum(y_desh_Trn_d- Trn_Out) / Trn_Int.shape[0] )
        derivative_1 = ( sum((y_desh_Trn_d- Trn_Out)*(Trn_Int)) / Trn_Int.shape[0] )
        
        Temp0= Thita[0,0] - learning_rate * derivative_0
        Temp1= Thita[1,0] - learning_rate * derivative_1
        
        new_Thita[0,0]=Temp0
        new_Thita[1,0]=Temp1
        
        Iterations += 1  
        print (Thita)
```

```python
Thita[0,0] = new_Thita[0,0]
Thita[1,0] = new_Thita[1,0]
y_desh_Trn_d = Thita[0,0] + (Thita[1,0] * Trn_Int) 
```

```python
plt.subplot(1,3,1)
plt.scatter(Trn_Int , Trn_Out )
plt.plot(Trn_Int , y_desh_Trn_d )
```

```python
MEAN_SE_Trn = np.dot((y_desh_Trn_d- Trn_Out),(y_desh_Trn_d- Trn_Out)) / (2 * Trn_Int.shape[0])
y_desh_Tst_d = Thita[0,0] + (Thita[1,0] * Tst_Int)
plt.subplot(1,3,3)
plt.scatter(Tst_Int , Tst_Out )
plt.plot(Tst_Int , y_desh_Tst_d)
plt.show()
MEAN_SE_Tst = np.sum((y_desh_Tst_d - Tst_Out)**2) / (2 * Tst_Int.shape[0])
```

```python
print ('Total no. of iteration for learning_rate: ' ,learning_rate  ,'are', Iterations)
print ("MEAN SQUARE ERROR on Training Data: " , MEAN_SE_Trn)
print ("MEAN SQUARE ERROR on Testing Data: " , MEAN_SE_Tst)
```
