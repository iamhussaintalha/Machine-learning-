# Linear Regression Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

trn_d= pd.read_csv('trainRegression.csv')
trn_d.head()
```

```python
trnx = np.array(trn_d['X']) 
trny = np.array(trn_d['R'])
```

```python
A = np.array([[len(trnx),sum(trnx)],[sum(trnx),sum(trnx**2)]])
A
```

```python
B = np.array([[sum(trny)],[sum(trny*trnx)]])
B
```

```python
A_inverse = np.linalg.inv(A)
A_inverse
```

```python
i= np.dot(A_inverse,B)
i
```

```python
Predtrn= i[0]+i[1]*trnx
plt.plot(trnx,trny)
plt.plot(trnx, Predtrn)
```

```python
#MSR FOT TRAINING DATA
sum(np.square(trny-Predtrn))/len(trnx)
```

```python
tst_d= pd.read_csv('testRegression.csv')
tstx = tst_d['X']
tsty = tst_d['R']
Predtst = i[0]+i[1]*tstx
plt.plot(tstx,tsty)
plt.plot(tstx,Predtst)
```

```python
#MSR FOR TESTING DATA
sum(np.square((tsty)-(Predtst)))/len(tstx)
```
