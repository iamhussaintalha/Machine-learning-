# SIXTH DEGREE REGRESSION MODEL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

trn_d= pd.read_csv('trainRegression.csv')
trn_d.head()
```

```python
trnx= np.array(trn_d['X']) 
trny= np.array(trn_d['R'])
```

```python
A=np.array([[len(trnx),sum(trnx),sum(trnx**2),sum(trnx**3),sum(trnx**4),sum(trnx**5),sum(trnx**6)],[sum(trnx),sum(trnx**2),sum(trnx**3),sum(trnx**4),sum(trnx**5),sum(trnx**6),sum(trnx**7)],[sum(trnx**2),sum(trnx**3),sum(trnx**4),sum(trnx**5),sum(trnx**6),sum(trnx**7),sum(trnx**8)],[sum(trnx**3),sum(trnx**4),sum(trnx**5),sum(trnx**6),sum(trnx**7),sum(trnx**8),sum(trnx**9)],[sum(trnx**4),sum(trnx**5),sum(trnx**6),sum(trnx**7),sum(trnx**8),sum(trnx**9),sum(trnx**9)],[sum(trnx**5),sum(trnx**6),sum(trnx**7),sum(trnx**8),sum(trnx**9),sum(trnx**10),sum(trnx**11)],[sum(trnx**6),sum(trnx**7),sum(trnx**8),sum(trnx**9),sum(trnx**10),sum(trnx**11),sum(trnx**12)]])
A
```

```python
B=np.array([[sum(trny)],[sum(trnx*trny)],[sum(trny*(trnx**2))],[sum(trny*(trnx**3))],[sum(trny*(trnx**4))],[sum(trny*(trnx**5))],[sum(trny*(trnx**6))]])
B
```

```python
A_inverse=np.linalg.inv(A)
A_inverse
```

```python
i=np.dot(A_inverse,B)
i
```

```python
Predtrn=i[0]+i[1]*trnx+i[2]*(trnx**2)+i[3]*(trnx**3)+i[4]*(trnx**4)+i[5]*(trnx**5)+i[6]*(trnx**6)
plt.plot(trnx,trny)
plt.plot(trnx,Predtrn)
```

```python
#MSR FOR TRAINING DATA
sum(np.square(trny-Predtrn))/len(trnx)
```

```python
tst_d= pd.read_csv('testRegression.csv')
tstx=tst_d['X']
tsty=tst_d['R']
```

```python
Predtst= i[0]+i[1]*tstx+i[2]*(tstx**2)+i[3]*(tstx**3)+i[4]*(tstx**4)+i[5]*(tstx**5)+i[6]*(tstx**6)
plt.plot(tstx,tsty)
plt.plot(tstx,Predtst)
```

```python
#MSR FOR TESTING DATA 
sum(np.square((tsty)-(Predtst)))/len(tstx)
```
