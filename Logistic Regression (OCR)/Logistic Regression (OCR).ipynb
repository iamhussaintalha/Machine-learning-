```python
import numpy as np
```

```python
Trn_X = np.loadtxt('trainX.txt', dtype=np.uint8)
C1s = np.ones((Trn_X.shape[0]), dtype=np.uint8)
Trn_X = np.insert(Trn_X , 0 , C1s , axis = 1)
```

```python
Trn_Y = np.loadtxt('trainY.txt', dtype=np.uint8)
Trn_Y = np.where(Trn_Y == 2 , 1 , 0)
```

```python
Tst_X = np.loadtxt('testX.txt', dtype=np.uint8)
C1s = np.ones((Tst_X.shape[0]), dtype=np.uint8)
Tst_X = np.insert(Tst_X , 0 , C1s , axis = 1)
```

```python
Tst_Y = np.loadtxt('testY.txt', dtype=np.uint8)
Tst_Y = np.where(Tst_Y == 2 , 1 , 0)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def cal_Acu(Pred , Act):
    return np.where(Pred == Act , 1 , 0).sum() / Act.shape[0] * 100
```

```python
learning_rates = [0.01, 0.1, 0.3, 0.5]

for learning_rate in learning_rates:
    Thita = np.random.uniform(0, 1, size=(Trn_X.shape[1]))
    i = 1
    m = Trn_X.shape[0]
    Pred = np.where(sigmoid(np.dot(Trn_X, Thita)) >= 0.5, 1, 0)
    Acu = cal_Acu(Pred, Trn_Y)

    while (not(Acu == 100) and (i <= 1000)):
        Prob = sigmoid(np.dot(Trn_X, Thita))
        Pred = np.where(Prob >= 0.5, 1, 0)
        Thita = np.array(Thita)
        Thita = Thita - (learning_rate / m) * np.dot(Trn_X.T, (Prob - Trn_Y))
        Acu = cal_Acu(Pred, Trn_Y)
        print('Learning Rate:', learning_rate, 'Accuracy for iteration:', i, ' is:', Acu)
        i += 1

    print('For learning rate:', learning_rate)
    print('Total no. of Iterations:', i - 1)
    Pred = np.where(sigmoid(np.dot(Trn_X, Thita)) >= 0.5, 1, 0)
    print('Accuracy on Training data:', cal_Acu(Pred, Trn_Y))
    Pred = np.where(sigmoid(np.dot(Tst_X, Thita)) >= 0.5, 1, 0)
    print('Accuracy on Testing data:', cal_Acu(Pred, Tst_Y))
```
